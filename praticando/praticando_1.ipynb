## ***Diagrama***


Cria√ß√£o Bando de dados

```python
CREATE DATABASE IF NOT EXISTS sloja_atelie;
```

Cria√ß√£o tabela categoria

```python
CREATE TABLE IF NOT EXISTS categoria(
    id_categoria INT AUTO_INCREMENT PRIMARY KEY,
    nome VARCHAR(50) NOT NULL
);
```

Cria√ß√£o tabela produto

```python
CREATE TABLE IF NOT EXISTS produto(
    sku INT NOT NULL AUTO_INCREMENT PRIMARY KEY, 
    nome VARCHAR(50) NOT NULL,
    valor DECIMAL(6,2),
    id_categoria INT NULL,
    FOREIGN KEY (id_categoria) REFERENCES categoria(id_categoria)
);
```

Cria√ß√£o tabela cliente

```python
CREATE TABLE IF NOT EXISTS cliente(
    id_cliente INT NOT NULL AUTO_INCREMENT PRIMARY KEY,
    nome VARCHAR(50) NOT NULL
);
```

Cria√ß√£o tabela pedido

```python
CREATE TABLE IF NOT EXISTS pedido(
    numero_pedido INT NOT NULL AUTO_INCREMENT PRIMARY KEY,
    id_cliente INT,
    sku_produto INT,
    quantidade INT,
    data DATETIME NOT NULL,
    FOREIGN KEY (sku_produto) REFERENCES produto(sku),
    FOREIGN KEY (id_cliente) REFERENCES cliente(id_cliente)
);
```

## ***Exibindo pedido com Jun√ß√£o***

```python
SELECT 
    pedido.numero_pedido,
    cliente.nome,
    produto.sku,
    produto.nome,
    categoria.nome,
    pedido.quantidade,
    pedido.data
    FROM cliente JOIN pedido
    ON pedido.id_cliente = cliente.id_cliente
    JOIN produto ON pedido.sku_produto = produto.sku
    JOIN categoria ON produto.id_categoria = categoria.id_categoria;
```
