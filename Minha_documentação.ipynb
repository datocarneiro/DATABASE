### ***ACID***
propriedades de transação conhecudas como ACID.

- Atomicidade: todas as operações em uma transação são tratadas como uma unidade indivisível: ou todas as operações são concluídas com sucesso, ou nenhuma delas é aplicada. Se uma operação falhar, a transação é revertida para o estado anterior.
- Consistência: a transação leva o banco de dados de um estado consistente para outro estado consistente. Isso significa que as restrições de integridade do banco de dados devem ser mantidas antes e após a execução da transação.
- Isolamento: cada transação é executada de forma isolada das outras. As modificações feitas por uma transação são invisíveis para outras transações até que a primeira seja concluída. Isso ajuda a evitar interferências e inconsistências causadas por operações concorrentes.
- Durabilidade: uma vez que uma transação é concluída com sucesso, suas alterações tornam-se permanentes no banco de dados, mesmo em caso de falha do sistema. Isso garante que as modificações sobrevivam a eventos como quedas de energia ou falhas no hardware.

`conceitos-chave.`

- Início de transação (BEGIN):
marca o início de uma transação. A partir desse ponto, as operações realizadas são tratadas como parte da transação até serem confirmadas ou revertidas.

- Confirmação (COMMIT): 
confirma todas as operações realizadas durante a transação. Isso torna as alterações permanentes no banco de dados.

- Reversão (ROLLBACK): 
desfaz todas as operações realizadas durante a transação. Isso restaura o banco de dados ao estado anterior à transação.
- Ponto de salvar (SAVEPOINT):
permite definir pontos intermediários dentro de uma transação. Isso facilita o ROLLBACK

- ROLLBACK para um estado específico em vez de reverter a transação inteira.

- Controle de concorrência: garante que transações concorrentes não causem problemas como leitura suja, leitura não repetível ou escrita fantasma. Mecanismos como bloqueios e isolamento transacional são empregados para gerenciar a concorrência.
- Log de transações: mantém o registro de todas as transações realizadas. Isso é crucial para a recuperação após falhas e para garantir a consistência do banco de dados.

### ***AS - como Variaveis***

É possível também utilizar a cláusula AS para a definição de `variáveis de registro`, variaveis para fazer referencia a uma tabela pelo alias.

Tais variáveis são `estabelecidas na cláusula FROM`, e ai no SELECT é possivel passar só o alias para referencia de qual tabela queremos trazer os registro.

Exemplo:
`variaveis para referenciar tabela "cidade" e "pais"`

NO exemplo estamos colocando a referencia para a tabela `city` em uma variavel definida como `C`.
e da mesma forma  estamos colocando a referencia para a tabela `country` em uma variavel definida como `CO`.

```sql
SELECT  C.Population AS PopulacaoCidade, CO.Population AS PopulacaoPaiz
FROM city AS C , country AS CO ;
```

então: `C.Population` irá consultar a população da `tabela city` 

então: `CO.População` irá consultar a população da `tabela PAIS`

## ***Operações de String e Ordenação nas consultas***

### ***LIKE (como) - tipo filtro***

`funciona como um filtro utilizando caracteres especiais`

tais como:

- Porcentagem (%): corresponde a qualquer substring.
- Sublinhado (_): corresponde a qualquer caracter

Sintax: registro LIKE "passa o filtro";

Exemplo:
`com %`

No exemplo abaixo a consulta irá retornar qualquer cidade que contem a palavra `"Curi"` e usamos `%` para informar que os demais caracteres posteriores a palavra CURI não importam


````python
SELECT * FROM city
WHERE Name LIKE "Curi%"; 
````

Exemplo:
`com _`

No exemplo abaixo a consulta irá retornar qualquer cidade que começãr com a letra `C` e ter o tamanho de 4 caracteres, pois no exemplo usamos uma sequencia de 4 Undercore


````python
SELECT * FROM city
WHERE Name LIKE "C___"; 
````

### ***Safe Update Mode - Modo seguro***

````python
Error Code: 
1175. You are using safe update mode and you tried to update a table without a WHERE that uses a KEY column.  
To disable safe mode, toggle the option in Preferences -> SQL Editor and reconnect.
````


O erro Error Code: 1175 ocorre quando você está tentando executar uma consulta de atualização (UPDATE) ou exclusão (DELETE) no MySQL enquanto o modo de atualização segura ("safe update mode") está ativado. Esse modo é ativado por padrão em algumas ferramentas MySQL (como o MySQL Workbench) para evitar alterações acidentais em grandes quantidades de registros sem uma condição WHERE específica.

`Para desativar o "Safe Update Mode" temporariamente:`

Você pode desativar o "Safe Update Mode" temporariamente na sessão atual do MySQL executando o seguinte comando:

````python
SET SQL_SAFE_UPDATES = 0;
````

Após isso, você pode executar sua consulta UPDATE ou DELETE normalmente. 
Quando terminar, se desejar, pode reativar o modo seguro:

`Para Reativar o "Safe Update Mode:`

````python
SET SQL_SAFE_UPDATES = 1;
````
