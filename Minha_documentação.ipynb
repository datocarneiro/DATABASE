Na linguagem SQL, destacam-se cinco subconjuntos de instrues



### ***ACID***
propriedades de transao conhecudas como ACID.

- Atomicidade: todas as operaes em uma transao so tratadas como uma unidade indivisvel: ou todas as operaes so concludas com sucesso, ou nenhuma delas  aplicada. Se uma operao falhar, a transao  revertida para o estado anterior.
- Consistncia: a transao leva o banco de dados de um estado consistente para outro estado consistente. Isso significa que as restries de integridade do banco de dados devem ser mantidas antes e aps a execuo da transao.
- Isolamento: cada transao  executada de forma isolada das outras. As modificaes feitas por uma transao so invisveis para outras transaes at que a primeira seja concluda. Isso ajuda a evitar interferncias e inconsistncias causadas por operaes concorrentes.
- Durabilidade: uma vez que uma transao  concluda com sucesso, suas alteraes tornam-se permanentes no banco de dados, mesmo em caso de falha do sistema. Isso garante que as modificaes sobrevivam a eventos como quedas de energia ou falhas no hardware.

`conceitos-chave.`

- Incio de transao (BEGIN):
marca o incio de uma transao. A partir desse ponto, as operaes realizadas so tratadas como parte da transao at serem confirmadas ou revertidas.

- Confirmao (COMMIT): 
confirma todas as operaes realizadas durante a transao. Isso torna as alteraes permanentes no banco de dados.

- Reverso (ROLLBACK): 
desfaz todas as operaes realizadas durante a transao. Isso restaura o banco de dados ao estado anterior  transao.
- Ponto de salvar (SAVEPOINT):
permite definir pontos intermedirios dentro de uma transao. Isso facilita o ROLLBACK

- ROLLBACK para um estado especfico em vez de reverter a transao inteira.

- Controle de concorrncia: garante que transaes concorrentes no causem problemas como leitura suja, leitura no repetvel ou escrita fantasma. Mecanismos como bloqueios e isolamento transacional so empregados para gerenciar a concorrncia.
- Log de transaes: mantm o registro de todas as transaes realizadas. Isso  crucial para a recuperao aps falhas e para garantir a consistncia do banco de dados.

## ***Operaes de String e Ordenao nas consultas***

### ***LIKE (como) - tipo filtro***

`funciona como um filtro utilizando caracteres especiais`

tais como:

- Porcentagem (%): corresponde a qualquer substring.
- Sublinhado (_): corresponde a qualquer caracter

Sintax: registro LIKE "passa o filtro";

Exemplo:
`com %`

No exemplo abaixo a consulta ir retornar qualquer cidade que contem a palavra `"Curi"` e usamos `%` para informar que os demais caracteres posteriores a palavra CURI no importam


````python
SELECT * FROM city
WHERE Name LIKE "Curi%"; 
````

Exemplo:
`com _`

No exemplo abaixo a consulta ir retornar qualquer cidade que comer com a letra `C` e ter o tamanho de 4 caracteres, pois no exemplo usamos uma sequencia de 4 Undercore


````python
SELECT * FROM city
WHERE Name LIKE "C___"; 
````

### ***Safe Update Mode - Modo seguro***

````python
Error Code: 
1175. You are using safe update mode and you tried to update a table without a WHERE that uses a KEY column.  
To disable safe mode, toggle the option in Preferences -> SQL Editor and reconnect.
````


O erro Error Code: 1175 ocorre quando voc est tentando executar uma consulta de atualizao (UPDATE) ou excluso (DELETE) no MySQL enquanto o modo de atualizao segura ("safe update mode") est ativado. Esse modo  ativado por padro em algumas ferramentas MySQL (como o MySQL Workbench) para evitar alteraes acidentais em grandes quantidades de registros sem uma condio WHERE especfica.

`Para desativar o "Safe Update Mode" temporariamente:`

Voc pode desativar o "Safe Update Mode" temporariamente na sesso atual do MySQL executando o seguinte comando:

````python
SET SQL_SAFE_UPDATES = 0;
````

Aps isso, voc pode executar sua consulta UPDATE ou DELETE normalmente. 
Quando terminar, se desejar, pode reativar o modo seguro:

`Para Reativar o "Safe Update Mode:`

````python
SET SQL_SAFE_UPDATES = 1;
````
